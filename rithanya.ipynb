```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io


df = pd.read_csv('happiness_2017.csv')
df_clean = df.dropna(subset=['HappinessScore', 'Log GDP per capita', 'Region'])

print("--- Descriptive Statistics ---")
print(df.describe())


plt.style.use('seaborn-v0_8')


plt.figure(figsize=(10, 6))
sns.histplot(df_clean['HappinessScore'], kde=True, color='skyblue')
plt.title('Distribution of Happiness Score (2017)')
plt.show()
top_10 = df_clean.sort_values(by='HappinessScore', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='HappinessScore', y='Country', data=top_10, palette='viridis')
plt.title('Top 10 Happiest Countries in 2017')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='Log GDP per capita', y='HappinessScore', hue='Region', data=df_clean)
plt.title('Happiness Score vs. Log GDP per Capita')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='HappinessScore', y='Region', data=df_clean, palette='Set3')
plt.title('Happiness Score Distribution by Region')
plt.show()

plt.figure(figsize=(12, 10))
numeric_df = df.select_dtypes(include=['float64', 'int64']).drop(columns=['Rank'], errors='ignore')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Happiness Factors')
plt.show()

key_vars = ['HappinessScore', 'Log GDP per capita', 'Social support', 'Healthy life expectancy at birth']
sns.pairplot(df_clean[key_vars])
plt.suptitle('Pair Plot of Key Happiness Indicators', y=1.02)
plt.show()
```
