```python
import numpy as np
import pandas as pd

sensor = np.array([22.1, 22.5, 23.0, 24.8, 25.1, 26.0, 27.5, 28.0])
s = pd.Series(sensor)

# Rolling mean and standard deviation
rolling_mean = s.rolling(window=3).mean()
rolling_std = s.rolling(window=3).std()

# Anomaly detection (|value - mean| > 1.5 * std)
anomalies = s[(abs(s - rolling_mean) > 1.5 * rolling_std)]
print("Rithanya 24BAD132")
print("Rolling Mean:\n", rolling_mean)
print("\nRolling Std:\n", rolling_std)
print("\nAnomalies:\n", anomalies)
```

```python
import numpy as np

interactions = np.array([
    [0, 0, 5, 0],
    [1, 0, 0, 0],
    [0, 0, 0, 3],
    [4, 0, 0, 0]
])

# Sparsity
zero_count = np.sum(interactions == 0)
sparsity = (zero_count / interactions.size) * 100

# Row normalization
row_sums = interactions.sum(axis=1, keepdims=True)
normalized = np.divide(interactions, row_sums, where=row_sums != 0)

# Top interaction per row
top_interactions = interactions.max(axis=1)
print("Rithanya 24BAD132")
print("Sparsity (%):", sparsity)
print("\nNormalized Matrix:\n", normalized)
print("\nTop Interaction per Row:", top_interactions)
```

```python
import pandas as pd

data = {
    "Student": ["A", "B", "C", "D", "E"],
    "Math": [78, 85, 67, 90, 72],
    "Physics": [74, 88, 69, 92, 70],
    "Attendance": [82, 90, 76, 95, 80]
}

df = pd.DataFrame(data)

# Average score
df["Average"] = df[["Math", "Physics", "Attendance"]].mean(axis=1)

# Grade assignment
def grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    else:
        return "C"

df["Grade"] = df["Average"].apply(grade)

# Top performer
top_performer = df.loc[df["Average"].idxmax()]
print("Rithanya 24BAD132")
print(df)
print("\nTop Performer:\n", top_performer)
```

```python
import pandas as pd
import numpy as np

timestamps = pd.date_range("2024-01-01", periods=10, freq="6H")
values = [12, 18, 20, 22, 19, 15, 14, 17, 21, 23]

df = pd.DataFrame({"Timestamp": timestamps, "Value": values})

df["Hour"] = df["Timestamp"].dt.hour
df["Day"] = df["Timestamp"].dt.day

# Cyclic encoding
df["Hour_sin"] = np.sin(2 * np.pi * df["Hour"] / 24)
df["Hour_cos"] = np.cos(2 * np.pi * df["Hour"] / 24)

peak_hour = df.loc[df["Value"].idxmax(), "Hour"]
print("Rithanya 24BAD132")
print(df)
print("\nPeak Activity Hour:", peak_hour)
```

```python
import pandas as pd

train = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "score": [65, 70, 80, 90],
    "result": [0, 0, 1, 1]
})

test = pd.DataFrame({
    "id": [3, 5],
    "score": [80, 75],
    "result": [1, 0]
})

# Overlapping IDs
overlap = set(train["id"]).intersection(set(test["id"]))

# Leakage percentage
leakage_percentage = (len(overlap) / len(test)) * 100
print("Rithanya 24BAD132")
print("Overlapping IDs:", overlap)
print("Potential Leakage Columns: id, result")
print("Leakage Percentage:", leakage_percentage)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": np.random.randint(0, 1000, 10000),
    "B": np.random.rand(10000),
    "C": ["category"] * 10000
})

before = df.memory_usage(deep=True).sum()

# Optimization
df["A"] = df["A"].astype("int16")
df["B"] = df["B"].astype("float32")
df["C"] = df["C"].astype("category")

after = df.memory_usage(deep=True).sum()
print("Rithanya 24BAD132")
print("Memory Before Optimization:", before)
print("Memory After Optimization:", after)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y = np.array([10, 12, 15, 18, 20, 22, 24, 23, 25, 27])
error = np.array([1, 1.2, 1.5, 1.3, 1.4, 1.6, 1.8, 1.7, 1.9, 2.0])
print("Rithanya 24BAD132")
plt.figure()
plt.plot(x, y)
plt.fill_between(x, y-error, y+error, alpha=0.3)

max_error_idx = np.argmax(error)
plt.annotate("Max Uncertainty",
             (x[max_error_idx], y[max_error_idx]))

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Uncertainty Visualization")
plt.show()
```
